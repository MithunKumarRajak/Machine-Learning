```python
data = pd.DataFrame({
    "age": [25, 30, None, 45, 50],
    "salary": [50000, 60000, 55000, None, 90000],
    "department": ["HR", "IT", "Finance", "IT", "HR"],
    "target": [0, 1, 0, 1, 1]
})
```

```python
data=sns.load_dataset('titanic')
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import *

```

```python
data.sample(5)
```

```python
si=SimpleImputer(strategy='mean')
data['age']=si.fit_transform(data[['age']])
```

```python
data.isnull().sum()
```

```python
si1=SimpleImputer(strategy='most_frequent')
data[['embarked','deck','embark_town']]=si1.fit_transform(data[['embarked','deck','embark_town']])
data.isnull().sum()
```
