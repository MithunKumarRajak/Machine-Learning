```python
import pandas as pd
```

```python
data={
    'Roll No': [1, 2, 3, 4], # Create a dictionary with data
    'Name': ['Shivam', 'Mithun', 'Sanket', 'Kamlesh'],
}
df=pd.DataFrame(data)
print(df)
```

```python
# df1=pd.DataFrame(data, index=['a', 'b', 'c', 'd']) # Create DataFrame with custom index
```

Accessing row with index with locate(loc)

```python

print(df.loc[1]) # Access row with index label 1
# Above Code returns a Pandas Series.
print("------------------------")
print(df.loc[[1]]) # Access row with index label 1 and 3
# Above code returns a Pandas DataFrame.
```

## With Custom index

```python
demoData = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(demoData, index = ["day1", "day2", "day3"])

print(df) 
```
