### Distribution Plots
Total 7 functions:

- displot() – flexible distribution plot (histogram, KDE, ECDF)

- histplot() – histogram (frequency distribution)

- kdeplot() – kernel density estimate (smooth curve)

- ecdfplot() – cumulative distribution function

- rugplot() – individual data points axis

- jointplot() – bivariate distribution + marginal distributions

- pairplot() – multiple variables ke pairwise distributions


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=sns.load_dataset('tips')
data.sample()
```

- distplot() ke jagah displot() use hota hai New version mai



```python
sns.distplot(data["total_bill"], bins=20, kde=True)
plt.show()
```

```python
sns.displot(data["tip"], bins=20, kde=True,color='m')
plt.show()
```

bins pure range ko utne peace mai divide karega then har ek division ke liye count karega 

kind="kde" --> smooth curve

- kind="kde" → smooth density curve banata hai.


```python
sns.displot(x='total_bill',data=data,kind='kde',fill=True,color='y')
```

- bins → number of intervals (default auto decide hota hai).
- kde=True → histogram ke upar smooth KDE curve bhi show karega.
- hue="sex" → categories ke basis par alag colors dikhayega


```python
sns.histplot(x='tip',data=data,bins=10,kde=True,hue='sex')
plt.show()
```
