### Distribution Plots
Total 7 functions:

- displot() – flexible distribution plot (histogram, KDE, ECDF)

- histplot() – histogram (frequency distribution)

- kdeplot() – kernel density estimate (smooth curve)

- ecdfplot() – cumulative distribution function

- rugplot() – individual data points axis

- jointplot() – bivariate distribution + marginal distributions

- pairplot() – multiple variables ke pairwise distributions


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=sns.load_dataset('tips')
data.sample()
```

- distplot() ke jagah displot() use hota hai New version mai



```python
sns.distplot(data["total_bill"], bins=20, kde=True)
plt.show()
```

```python
sns.displot(data["tip"], bins=20, kde=True,color='m')
plt.show()
```

bins pure range ko utne peace mai divide karega then har ek division ke liye count karega 

kind="kde" --> smooth curve

- kind="kde" → smooth density curve banata hai.


```python
sns.displot(x='total_bill',data=data,kind='kde',fill=True,color='y')
```

- bins → number of intervals (default auto decide hota hai).
- kde=True → histogram ke upar smooth KDE curve bhi show karega.
- hue="sex" → categories ke basis par alag colors dikhayega


```python
sns.histplot(x='tip',data=data,bins=10,kde=True,hue='sex')
plt.show()
```

```python
sns.histplot(data=data[['tip','total_bill']],kde=True,bins=20)
```

or

```python
sns.histplot(x="tip", data=data, kde=True, color="blue", label="Tip")
sns.histplot(x="total_bill", data=data, kde=True, color="red", label="Total Bill")
plt.show()
```

- kdeplot()

kdeplot() --> smooth distribution curve banata hai. Ye histogram ka alternative hai

cumulative=True se curve cumulative banta hai


```python
sns.kdeplot(x='tip',data=data, fill=True, color= 'b')
```

```python
sns.kdeplot(x="total_bill", y="tip", data=data, fill=True)
```

- jointplot()  --> bivariate relationship

kind='__' mai jo likhoge usme uss type ka graph banega

kind="scatter"

kind="kde"

kind="hex"

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='scatter')
plt.show()
```

- Center → scatter plot (bill vs tip relation).
- Top → total_bill ka histogram.
- Right → tip ka histogram.


```python
sns.jointplot(x="total_bill", y="tip", data=data, kind="hex")
plt.show()
```

- pointplot() -->

```python
sns.pointplot(x="day", y="tip", data=data)
plt.show()
```

linestyle == join

```python
sns.pointplot(x="day", y="tip", hue="sex", data=data,linestyle='none',ci=50)
plt.show()
```

- rugplot() --> individual data points

- Ye data ke spread aur density ko quickly visualize karne ke liye use hota hai


```python
sns.rugplot(y='tip',data=data,height=0.8)
```

```python
sns.kdeplot(x="tip", data=data, fill=True)
sns.rugplot(x="tip", data=data)
plt.show()
```

- ecdfplot()

- ecdfplot() --> ye batata hai ki ek value tak kitna % data aa chuka hai.


```python
sns.ecdfplot(x='tip',data=data)
plt.show
```

```python
sns.ecdfplot(x='tip',data=data,hue='smoker')
plt.show
```
