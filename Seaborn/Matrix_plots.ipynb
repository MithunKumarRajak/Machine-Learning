```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=sns.load_dataset('tips')
data.sample(2)
```

- heatmap()

- annot=True → values show karo.
- annot=False (default) → sirf colors show hote hain, numbers nahi.


linewidths
- Har cell ke beech line ki thickness control karta hai.

linecolor
- Grid lines ka color set karta hai.


cmap (Colormap)
- Color scheme control karta hai


- linewidths + linecolor → grid lines control karte hain.
- square → cells ko square banata hai.
- cmap → colors change karta hai.
- mask → kuch cells hide karne ke liye.



```python
corr = data.select_dtypes(include=['float64','int64']).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm",linewidths=1,linecolor="black")
plt.show()
```

```python
# Load example dataset
flights = sns.load_dataset("flights")
# Pivot the data to get a matrix form (Month vs Year)
flights_matrix = flights.pivot(index="month", columns="year", values="passengers")

# Basic Heatmap
sns.heatmap(flights_matrix)
plt.show()
```

```python
data_pivot=data.pivot_table(index="day", columns="sex", values="tip")
sns.heatmap(data_pivot)
```

### clustermap()

- clustermap() ek advanced heatmap hai jo clustering + dendrogram add karta hai.
- Ye automatically rows/columns ko group karta hai similarity ke basis par.
- Useful hai jab tumhe patterns aur hidden relationships samajhne ho.


```python
corr = data.select_dtypes(include=['float64','int64']).corr()

sns.clustermap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
data_pivot=data.pivot_table(index="day", columns="sex", values="total_bill")
sns.clustermap(data_pivot)
```
