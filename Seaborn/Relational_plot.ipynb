### Relational plots

Used for visualizing the statistical relationship between the data points

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.get_dataset_names()
dataset=sns.load_dataset('tips')
dataset.sample()
```

To draw the relational plots seaborn provides 3 functions.

relplot()

scatterplot()

lineplot()

```python
sns.relplot(x='total_bill',y='tip',data=dataset)
```

hue='sex'

Grouping variable with different colors.

```python
sns.relplot(x='total_bill',y='tip',data=dataset,hue='time')
```

```python
sns.scatterplot(x='tip',y='total_bill',data=dataset,hue='smoker')
```

```python
sns.lineplot(x='total_bill',y='tip',data=dataset,hue='smoker')
```
