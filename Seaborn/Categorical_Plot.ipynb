### Categorical Plots
category like a group, class or division.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Distribution-Based Plots

barplot() - Compare averages

boxplot() - Show distribution
 
countplot() - Compare counts

violinplot() - Show distribution

stripplot() - Show raw observations

swarmplot() - Show raw observations

boxenplot() - Large dataset distribution


* catplot() â†’ can draw all categorical plots.

```python
data=sns.load_dataset('tips')
data.head()
```

- Catplot()

sns.catplot(x,y,data,kind='__')

```python
sns.catplot(x='day',y='tip',data=data,kind='box',hue ='sex')
plt.show()
```

* Barplot()

To use this plot we choose a categorical column for the x axis and a numerical column for the y axis and we see that it creates a plot taking a mean per categorical column.

```python
sns.barplot(x='day',y='tip',data=data)
plt.show()
```

- Agar tumhe frequency dekhni hai to Seaborn mein countplot()


```python
sns.countplot(x='smoker',data=data)
plt.show()
```

boxplot()

IQR , OUTLIRE

```python
sns.boxplot(x='day',y='tip',data=data)
plt.show()
```

violine plot = box plot + kde plot 

for outlier : 

sns.violin(x,y,data,inner='point')


```python
sns.violinplot(x='day',y='tip',data=data,inner='point')
plt.show()
```

- stripplot()

for individual points 

sns.stripplot(x,y,data,jitter=True) -> jitter=True for don't overlap

easlly outlier finded


```python
sns.stripplot(x='day',y='tip',data=data , jitter=True)
plt.show()
```

- swarmplot()

like strip plot but dots ko systematically arrange karta hai

easily outlier detected

```python
sns.swarmplot(x='day',y='tip',data=data)
plt.show()
```

catplot()

all in one plot , with kind='__' parameter

```python
sns.catplot(x="day", y="tip", kind="strip", hue="sex", data=data)
plt.show()
```
