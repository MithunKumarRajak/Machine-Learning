### Subplot
To Display Multiple Plots

With the subplot() function you can draw multiple plots in one figure:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot 1
x = np.array([10, 11, 2, 31])
y = np.array([35, 81, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([10, 18, 2, 43])
y = np.array([10, 20, 30, 40])


plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1
x = np.array([0, 14, 25, 31])
y = np.array([3, 8, 14, 10])

plt.subplot(1, 3, 1)
plt.plot(x,y)
plt.title("First Plot")

#plot 2
x = np.array([0, 11, 21, 35])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 3, 2)
plt.plot(x,y)
plt.title("Second Plot",loc='right')

# plat 3
x = np.array([10, 11, 12, 13])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 3, 3)
plt.title("Third Plot")
plt.plot(x,y)

plt.suptitle("Super Title")
plt.show()
```
