### Plotting x and y points

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis

```python
# point on x axis = 1,1 and point on y axis = 5,1
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,5]) # Both x coordinates from both points
y=np.array([1,1]) # Both y coordinates from both points
plt.plot(x,y)
plt.show()
```


Above x and y point :

Both x coordinates from both points in x

Both y coordinates from both points in y

```python
# point 1,1 and 2,2
x=np.array([1,2])
y=np.array([1,2])
plt.plot(x,y,'o') # 'o' for markers only 
plt.show()
```

### Multiple Points
But make sure, have the same number of points in both axis.

```python
# point 1,1 2,1 3,1 4,1 5,1 and 1,2 2,2 3,2 4,2 5,2
x=np.array([1,2,3,4,5])
y=np.array([1,1,1,1,1])
plt.plot(x,y,'o') # 'o' for markers only
plt.show()
```

## Default X-Points

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.


```python
y = np.array([0,1,3.5,2,5,3]) # Here x is not defined, so by default x = [0,1,2,3,4,5]
plt.plot(y)
plt.show()
```
