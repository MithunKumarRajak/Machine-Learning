### Bars Graph
With Pyplot, you can use the bar() function to draw bar graphs:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array(['A','B','C','D','E'])
y=np.array([5,7,2,4,6])

plt.bar(x,y)
plt.show()
```

The bar() function takes arguments

categories and their values represented by the first and second argument as arrays.

* If you want the bars to be displayed horizontally instead of vertically, use the 'barh()' function

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 41, 10])

plt.barh(x, y)
plt.show()
```

### Bar Color
The 'bar()' and 'barh()' take the keyword argument 'color' to set the color of the bars

plt.bar(x, y, color = "green")


```python
x = np.random.rand(5)
y = np.random.rand(5)

plt.barh(x, y, color='g')
plt.show()
```

### Bar Width
The 'bar()' takes the keyword argument width to set the 'width' of the bars

width=0.2

```python
x1 = np.random.rand(5)
y1 = np.random.rand(5)

plt.bar(x1, y1, color='r',width=0.01)
```

The default width value is 0.8

Note: For horizontal bars, use 'height' instead of width

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([13, 58, 15, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```
