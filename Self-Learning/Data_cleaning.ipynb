Empty values, or Null values, can be bad when analyzing data,

and we should consider removing rows with empty values.

this is a step towards what is called cleaning data



# Cleaning Empty Cells


```python
import pandas as pd
print(pd.__version__)
```

```python
df=pd.read_csv('DataSet/fitness_data.csv')
new_df=df.dropna()
print(df)
print(new_df.to_string())
```

Note: By default, the dropna() method returns a new DataFrame, and will not change the original.

If you want to change the original DataFrame, use the inplace = True argument:

The dropna(inplace = True) will NOT return a new DataFrame, but it will remove all rows containing NULL values from the original DataFrame.

### dropna() index ke sath hi row ko delete kar deta hai


# #Filling empty cell with values

```python
df.fillna(130, inplace=True)
print(df.to_string())
```

### Replace data Using Mean, Median, or Mode


```python
x = df['Pulse'].mean()
new_df = df['Pulse'].fillna(x)
print(new_df.to_string())
```

```python
x = df["Calories"].mode()[0]
df["Calories"].fillna(x, inplace=True)
print(df.to_string())
```
